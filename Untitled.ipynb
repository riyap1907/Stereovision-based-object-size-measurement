```python
#Importing required libraries
import pyrealsense2 as rs
import numpy as np
import cv2
import glob
import random
import os
import re
net = cv2.dnn.readNet('best.onnx')
classes = ['biker', 'car', 'pedestrian', 'trafficLight', 'trafficLight-Green', 'trafficLight-GreenLeft', 'trafficLight-Red', 'trafficLight-RedLeft', 'trafficLight-Yellow', 'trafficLight-YellowLeft', 'truck']
while True:
    
    #Get frames from camera
    color_frame = "ped.png"
    images_path = [color_frame]
    
    #Get layers from neural network
    layer_names = net.getLayerNames()
    output_layers = [layer_names[i - 1] for i in net.getUnconnectedOutLayers()]
    
    #Get color codes in group of 3 containing random values from 0 to 255
    colors = np.random.uniform(0, 255, size=(len(classes), 3))

    for img_path in images_path:
        
                #Load color frames
                img = img_path
                
                #Resize frames
                img = cv2.resize(img, None, fx=0.4, fy=0.4)
                height, width, channels = img.shape

                #Preprocess frames
                blob = cv2.dnn.blobFromImage(img, 0.00392, (416, 416), (0, 0, 0), True, crop=False)

                #Detect objects
                net.setInput(blob)
                outs = net.forward(output_layers)


                class_ids = []
                confidences = []
                boxes = []
                
                for out in outs:
                    
                    for detection in out:
                        
                        #Obtain scores, class IDs and confidence levels of frames
                        scores = detection[5:]
                        class_id = np.argmax(scores)
                        confidence = scores[class_id]
                        
                        #Detection of object
                        if confidence > 0.5:
                            
                            print(class_id)
                            #Get centre coordinates, width and height
                            center_x = int(detection[0] * width)
                            center_y = int(detection[1] * height)
                            w = int(detection[2] * width)
                            h = int(detection[3] * height)

                            #Get rectangle coordinates
                            x = int(center_x - w / 2)
                            y = int(center_y - h / 2)

                            #Store box dimensions, confidence levels and class IDs
                            boxes.append([x, y, w, h])
                            confidences.append(float(confidence))
                            class_ids.append(class_id)
                            
                            #Get the distance from camera using depth frame
#                             distance = depth_frame[center_y,center_x]/1000
#                             print(distance)
                            cv2.circle(img,(center_x,center_y),1,(255, 0, 0),-1)
#                             cv2.putText(img, "{}m".format(distance), (center_x, center_y - 20), cv2.FONT_HERSHEY_PLAIN, 2, (255, 0, 0), 2)

                #Non Maximum Supression of boxes using confidence levels
                indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)
                print(indexes)
                
                font = cv2.FONT_HERSHEY_PLAIN
                for i in range(len(boxes)):
                    
                    if i in indexes:
                        
                        #Draw boxes with labels around objects
                        x, y, w, h = boxes[i]
                        label = str(classes[class_ids[i]])
                        color = colors[class_ids[i]]
                        cv2.rectangle(img, (x, y), (x + w, y + h), color, 2)
                        
    #Show color frame
    cv2.imshow("Color frame", img)
    key = cv2.waitKey(1)
    if key == 27:
        break
```
